```python
import numpy as np
import matplotlib.pyplot as plt
import iPython.display as ipd
```

```python
# parameter setting
amp = 1
sr = 10000
dur = 0.5
freq = 100.0
phase = 0
```

```python
# generate sample ticks
nSamp = sr * dur
samples = np.arange(1, nSamp+1)
```

```python
# convet samples to time
t = samples / sr
```

```python
# conver time to angle in range
theta0 = t * 2*np.pi * freq
```

```python
# shift angle by phase
theta = theta0 + phase
```

```python
# generate signal by complex-phasor
c = amp*np.exp(theta*1j)
```

```python
# plot the entire signal
import plotly.graph_objects as go
x, y, z = t, c.real, c.imag
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z,
                                   mode = 'markers', 
                                   marker = dict(size=3))])
fig.update_layout(scene = dict(
                    xaxis_title = 'time (sec)',
                    yaxis_title = 'real',
                    zaxis_title = 'imag'))
fig.show()
```

```python
# plot a certain portion of signals
import plotly.graph_objects as go
sampleRng = np.arange(0,100)
x, y, z = t[sampleRng], c.real[sampleRng], c.imag[sampleRng]
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z,
                                   mode='markers',
                                   marker=dict(size=3))])
fig.update_layout(scene = dict(
                    xaxis_title='time (sec)',
                    yaxis_title='real',
                    zaxis_title='imag'))
fig.show()
```
